```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("keremkarayaz/coffee-shop-sales")

print("Path to dataset files:", path)
```

```python
# Load data
import pandas as pd

df = pd.read_excel(path + "/Coffee Shop Sales.xlsx", sheet_name="Transactions")
df
```

```python
# Check Null Values
df.isnull().sum()
```

```python
# Check Duplicated value
df.duplicated().sum()
```

```python
# Transformations
# Convert Transaction Date

df['Date'] = pd.to_datetime(df['transaction_date'])
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day


```

```python
# 6. Convert Transaction Time to datetime & extract

df['Time'] = pd.to_datetime(df['transaction_time'].astype(str))
df['Hour'] = df['Time'].dt.hour
df['Minute'] = df['Time'].dt.minute
```

```python
df
```

```python
# ===== Transform Categorical Columns to Numeric =====
# product_category mapping
category_map = {
    "Coffee": 1,
    "Tea": 2,
    "Bakery": 3,
    "Drinking Chocolate": 4,
    "Flavours": 5
}
df["product_category_num"] = df["product_category"].map(category_map)
```

```python
# store_location mapping
location_map = {
    "Hell's Kitchen": 1,
    "Astoria": 2,
    "Lower Manhattan": 3
}
df["store_location_num"] = df["store_location"].map(location_map)
```

```python
# product_type mapping
product_type_map = {
    "Brewed Chai tea": 1,
    "Gourmet brewed coffee": 2,
    "Barista Espresso": 3,
    "Hot chocolate": 4,
    "Brewed Black tea": 5,
    "Drip coffee" : 6
}
df["product_type_num"] = df["product_type"].map(product_type_map)
```

```python
# Show transformed data
print(df[["product_category","product_category_num",
          "store_location","store_location_num",
          "product_type","product_type_num"]].head())
```

```python
df["product_category_num"] = df["product_category_num"].fillna(0).astype(int)
df["store_location_num"] = df["store_location_num"].fillna(0).astype(int)
df["product_type_num"] = df["product_type_num"].fillna(0).astype(int)
```

```python
print(df[["product_category","product_category_num",
          "store_location","store_location_num",
          "product_type","product_type_num"]].head())
```

```python
# Total Amount

df['total_amount'] = df['transaction_qty'] * df['unit_price']
```

**Star Schema**

```python
# Date Dimension
dim_date = df[['transaction_date','Date','Year','Month','Day','Hour','Minute']].drop_duplicates().reset_index(drop=True)
dim_date['Date_Key'] = range(1, len(dim_date) + 1)
```

```python
# Store Dimension
dim_store = df[['store_id','store_location','store_location_num']].drop_duplicates().reset_index(drop=True)
dim_store['Store_Key'] = range(1, len(dim_store) + 1)
```

```python
# Product Dimension
dim_product = df[['product_id','product_category','product_category_num','product_type','product_type_num']].drop_duplicates().reset_index(drop=True)
dim_product['Product_Key'] = range(1, len(dim_product) + 1)
```

```python
# (Fact table )
fact_transactions = df[['transaction_id','store_id','product_id','Date','transaction_qty','unit_price','total_amount','product_id']].copy()
fact_transactions['Transaction_Key'] = range(1, len(fact_transactions) + 1)
```

```python
# Save Them As CSV Files
fact_transactions.to_csv("fact_transactions.csv",index=False)
dim_date.to_csv("dim_date.csv",index=False)
dim_product.to_csv("dim_product.csv",index=False)
dim_store.to_csv("dim_store.csv",index=False)
```
